```python
# Import relevant libraries
from splinter import Browser
from bs4 import BeautifulSoup as soup
import matplotlib.pyplot as plt
import pandas as pd
```

```python
browser = Browser('chrome')
```

### Step 1: Visit the Website

Use automated browsing to visit the [Mars Temperature Data Site](https://static.bc-edx.com/data/web/mars_facts/temperature.html). Inspect the page to identify which elements to scrape.

   > **Hint** To identify which elements to scrape, you might want to inspect the page by using Chrome DevTools to discover whether the table contains usable classes.


```python
# Visit the website
# https://static.bc-edx.com/data/web/mars_facts/temperature.html
url = "https://static.bc-edx.com/data/web/mars_facts/temperature.html"
browser.visit(url)
```

### Step 2: Scrape the Table

Create a Beautiful Soup object and use it to scrape the data in the HTML table.

Note that this can also be achieved by using the Pandas `read_html` function. However, use Beautiful Soup here to continue sharpening your web scraping skills.

```python
# Create Beautiful Soup Object, Parse HTML
html = browser.html
soup = soup(html, 'html.parser')
```

```python
# Extract all rows of data
# Extract All Rows Of Data
all_data_rows = soup.find_all(class_='data-row')
all_data_rows
```

### Step 3: Store the Data

Assemble the scraped data into a Pandas DataFrame. The columns should have the same headings as the table on the website. Here’s an explanation of the column headings:

* `id`: the identification number of a single transmission from the Curiosity rover
* `terrestrial_date`: the date on Earth
* `sol`: the number of elapsed sols (Martian days) since Curiosity landed on Mars
* `ls`: the solar longitude
* `month`: the Martian month
* `min_temp`: the minimum temperature, in Celsius, of a single Martian day (sol)
* `pressure`: The atmospheric pressure at Curiosity's location

```python
# Create An Empty List
mars_temperature_data = []

# Loop Through The Scraped Data To Create A List Of Rows Of Data
for row in all_data_rows:
    
    # Find All Table Cells In The Row
    cells = row.find_all('td')
    
    # Extract Text From Each Cell And Create A Row
    row_data = [cell.get_text() for cell in cells]
    
    # Append The Row To The List 'mars_temperature_data'
    mars_temperature_data.append(row_data)

# 'mars_temperature_data' List Contains All Rows Created From The Scraped Data
print(mars_temperature_data)
```

```python
# Create a Pandas DataFrame by using the list of rows and a list of the column names
# Create Pandas DataFrame By Using List Of Rows And List Of Column Names
mars_temperature_data_columns = ['id','terrestrial_date','sol','ls','month','min_temp','pressure']
mars_temperature_data_df = pd.DataFrame(mars_temperature_data, columns = mars_temperature_data_columns)
```

```python
# Confirm DataFrame was created successfully
mars_temperature_data_df.head()
```

### Step 4: Prepare Data for Analysis

Examine the data types that are currently associated with each column. If necessary, cast (or convert) the data to the appropriate `datetime`, `int`, or `float` data types.

  > **Hint** You can use the Pandas `astype` and `to_datetime` methods to accomplish this task.


```python
# Examine data type of each column
mars_temperature_data_df.info()
```

```python
# Change data types for data analysis
mars_temperature_data_df['terrestrial_date'] = pd.to_datetime(mars_temperature_data_df['terrestrial_date'])
mars_temperature_data_df[['sol','ls','month']] = mars_temperature_data_df[['sol','ls','month']].astype('int64')
mars_temperature_data_df[['min_temp','pressure']] = mars_temperature_data_df[['min_temp','pressure']].astype('float64')
```

```python
# Confirm type changes were successful by examining data types again
mars_temperature_data_df.info()
```

### Step 5: Analyze the Data

Analyze your dataset by using Pandas functions to answer the following questions:

1. How many months exist on Mars?
2. How many Martian (and not Earth) days worth of data exist in the scraped dataset?
3. What are the coldest and the warmest months on Mars (at the location of Curiosity)? To answer this question:
    * Find the average the minimum daily temperature for all of the months.
    * Plot the results as a bar chart.
4. Which months have the lowest and the highest atmospheric pressure on Mars? To answer this question:
    * Find the average the daily atmospheric pressure of all the months.
    * Plot the results as a bar chart.
5. About how many terrestrial (Earth) days exist in a Martian year? To answer this question:
    * Consider how many days elapse on Earth in the time that Mars circles the Sun once.
    * Visually estimate the result by plotting the daily minimum temperature.


```python
# 1. How many months are there on Mars?
# Count the number of unique months in the 'month' column of the Mars Temperature DataFrame
number_of_months_mars = mars_temperature_data_df['month'].nunique()

# Display the result
print("Number of Months Present on Mars:", number_of_months_mars)
```

```python
# 2. How many Martian days' worth of data are there?
# Calculate Total Number Of Martian Days By Counting Length Of 'sol' Column In Mars Temperature DataFrame
total_days_mars_data = len(mars_temperature_data_df['sol'])

# Print The Result
print("Number of Martian Days' Worth of Data:", total_days_mars_data)
```

```python
# 3. Calculate the average low temperature for each month on Mars.
# Group the Mars Temperature DataFrame by the 'month' column and compute the mean of the 'min_temp' column.
average_low_temperature_by_month = mars_temperature_data_df.groupby('month')['min_temp'].mean()

# Display the results.
average_low_temperature_by_month
```

```python
# Plot the average temperature by month
average_low_temperature_by_month.plot(kind='bar')

# Set Labels And Title
plt.xlabel('\nMonth')
plt.ylabel('Temperature °C\n')
plt.title('Average Low Temperature by Month on Mars\n')

# Show Plot
plt.show()
```

```python
# Identify the coldest and hottest months in Curiosity's location
average_low_temperature_by_month.sort_values().plot(kind='bar')

# Set Labels And Title
plt.xlabel('\nMonth')
plt.ylabel('Temperature °C\n')
plt.title('Average Low Temperature by Month on Mars\n')

# Show Plot
plt.show()
```

```python
# 4. Average pressure by Martian month
average_pressure_by_month = mars_temperature_data_df.groupby('month')['pressure'].mean()

# Display Results
average_pressure_by_month
```

```python
# Plot the average pressure by month
average_pressure_by_month.sort_values().plot(kind='bar')

# Set Labels And Title
plt.xlabel('\nMonth')
plt.ylabel('Atmospheric Pressure\n')
plt.title('Average Atmospheric Pressure by Month on Mars\n')

# Show Plot
plt.show()
```

```python
# 5. How many terrestrial (earth) days are there in a Martian year?
# Filter Mars Temperature DataFrame To Include Only Rows Where Solar Longitude Is Either 0 Or 359
mars_orbit_data = mars_temperature_data_df[mars_temperature_data_df['ls'].isin([0, 359])]

# Sort Filtered DataFrame By 'terrestrial_date' Column
mars_orbit_sorted = mars_orbit_data.sort_values(by='terrestrial_date')

# Group Sorted Data By Solar Longitude And Terrestrial Date, Count Occurrences
mars_orbit_grouped = mars_orbit_sorted.groupby(['ls', 'terrestrial_date']).size()

# Display Resulting Series
mars_orbit_grouped
```

```python
# Calculation For Difference In Terrestrial (Earth) Days Corresponding To One Orbit Of Mars Around Sun

# Define Two Dates As Strings, Corresponding To Solar Longitude 0 And 359, Respectively
date_ls_orbit_start = '2013-08-01'
date_ls_orbit_end = '2015-06-18'

# Convert The Date Strings To Pandas Datetime Objects
date_start = pd.to_datetime(date_ls_orbit_start)
date_end = pd.to_datetime(date_ls_orbit_end)

# Calculate The Difference
date_difference = date_end - date_start

# Print The Result
print("Datetime Difference:", date_difference)
```

```python
# Plot Minimum Temperature Against Number Of Martian Days (sol)
plt.plot(mars_temperature_data_df['sol'], mars_temperature_data_df['min_temp'])

# Add Labels And Title
plt.xlabel('\nNumber of Terrestrial Days')
plt.ylabel('Minimum Temperature\n')
plt.title('Number of Terrestrial Days vs Minimum Temperature\n')

# Show The Plot
plt.show()
```

On average, the third month has the coldest minimum temperature on Mars, and the eighth month is the warmest. But it is always very cold there in human terms!



Atmospheric pressure is, on average, lowest in the sixth month and highest in the ninth.

The distance from peak to peak is roughly 1425-750, or 675 days. A year on Mars appears to be about 675 days from the plot. Internet search confirms that a Mars year is equivalent to 687 earth days.
